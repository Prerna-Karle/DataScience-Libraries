```python
#Importing the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Importing the dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Setting a theme for the plots
sns.set_theme()
```

```python
#Visualize the data
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col = 'time', style = 'smoker', size = 'size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
#scatter plot using different parameters
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```
